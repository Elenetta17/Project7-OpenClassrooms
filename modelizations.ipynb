# **Implémentez un modèle de scoring**

La société financière "Prêt à dépenser" propose des crédits à la consommation pour des personnes ayant peu ou pas du tout d'historique de prêt.

L’entreprise souhaite mettre en œuvre un outil de “scoring crédit” pour calculer la probabilité qu’un client rembourse son crédit, puis classifie la demande en crédit accordé ou refusé. Elle souhaite donc développer un algorithme de classification en s’appuyant sur des sources de données variées (données comportementales, données provenant d'autres institutions financières, etc.).

De plus, les chargés de relation client ont fait remonter le fait que les clients sont de plus en plus demandeurs de transparence vis-à-vis des décisions d’octroi de crédit. Cette demande de transparence des clients va tout à fait dans le sens des valeurs que l’entreprise veut incarner.

Prêt à dépenser a décidé donc de développer un dashboard interactif pour que les chargés de relation client puissent à la fois expliquer de façon la plus transparente possible les décisions d’octroi de crédit, mais également permettre à leurs clients de disposer de leurs informations personnelles et de les explorer facilement. 

Mon objectif est donc:
- Construire un modèle de scoring qui donnera une prédiction sur la probabilité de faillite d'un client de façon automatique.
- Construire un dashboard interactif à destination des gestionnaires de la relation client permettant d'interpréter les prédictions faites par le modèle, et d’améliorer la connaissance client des chargés de relation client.
- Mettre en production le modèle de scoring de prédiction à l’aide d’une API, ainsi que le dashboard interactif qui appelle l’API pour les prédictions.

Ce notebook est consacré à la modélisation. La démarche de nettoyage du jeu de données et l'analyse exploratoire des données dont disponibles dans le notebook **EDA.ipynb**.

# 2. Les metriques

Je vais évaluer les modèles à l'aide de plusieurs métriques :
- ROC-AUC, qui mesure la capacité d'un modèle à distinguer les deux classes
- accuracy, qui permet de voir combien de clients sont classés correctement
- recall, qui permet de savoir le pourcentage de positifs (défaut) bien prédits
- une métrique métier (credit-score) qui consiste dans une somme pondérée et normalisée de faux négatifs et faux positifs. Le faux négatifs (clients auxquels on accorde un prêt mais qui vont faire défaut) auront un poids de 10, car ils constituent une perte financière pour l'entreprise. Les faux positifs auront un poids de 1, car même s’ils ne représentent pas directement une perte d’argent, ils représentent un manque de gain.

```python
# custom metric
def custom_score(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    return (10*cm[1, 0] + cm[0, 1])/y_true.shape[0]


# saving the metric
credit_score = make_scorer(custom_score, greater_is_better=False)
sklearn.metrics.SCORERS['credit_score'] = credit_score
```

```python
# function to compute and show the scores
def scores(y_true, probabilities):
    roc_auc = roc_auc_score(y_true, probabilities)
    classes = [1 if proba > 0.5 else 0 for proba in probabilities]
    accuracy = accuracy_score(y_true, classes)
    recall = recall_score(y_true, classes)
    custom = custom_score(y_true, classes)
    print("ROC-AUC: {}".format(roc_auc))
    print("Accuracy: {}".format(accuracy))
    print("Recall: {}".format(recall))
    print("Credit_score: {}".format(custom))
    return [roc_auc, accuracy, recall, custom]
```
